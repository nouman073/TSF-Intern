The dataset contains data 12 years, from 2008 to 2019.

**Analysis based on teams**

```python
df.winner.value_counts()
```

```python
data=df.winner.value_counts().sort_index()
sns.countplot(data=df,y='winner')
```

```python
df.groupby(by= ["winner"])['city'].value_counts().sort_values(ascending = False).head()
```

Over the last 12 years, the team with the highest number of victories is Mumbai Indians.

**Analysis based Hosted city**

```python
data=df.city.value_counts()
```

```python
data = df.city.value_counts()
sns.barplot(y = data.index, x = data)
plt.title('Number of matches hosted')
```

After analyzing the bar plot, it is evident that Mumbai has emerged as the top host city for
 IPL matches, having the highest number of matches hosted.

**Analysis based on the toss decision**

```python
df.toss_decision.value_counts()
```

```python
sns.countplot(df, x='toss_decision')
plt.title("Mostly teams prefer the field first")
```

```python
toss_win_match=df[(df['toss_winner'] == df['winner'])]
toss_win_lose_match=df[(df['toss_winner'] != df['winner'])]
```

```python
toss_df = pd.DataFrame({"Team won the toss":[ toss_win_match.shape[0],toss_win_lose_match.shape[0]  ]},
                       index = ['Match won', 'Match Lost']  )

toss_df
```

```python
sns.barplot(x=toss_df.index, y="Team won the toss", data=toss_df)
plt.title("Number of matches where the team won the toss and the match result")
```

The analysis reveals that the toss winner does not have a substantial impact on the outcome of matches. However, it is worth noting that the number of matches won is slightly higher than the number of matches lost.

**Analysis based on Player of match**

```python
data=df.player_of_match.value_counts().head()
```

```python
sns.barplot(y = data.index, x = data)

```

The above results indicate that Chris Gayle and AB de Villiers have received the "Player of the Match" award most frequently. This observation suggests that teams may benefit from considering the inclusion of these players in their squad during auctions or player acquisitions.
