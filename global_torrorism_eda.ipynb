```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('globalterrorismdb_0718dist.csv',encoding='latin1')
```

```python
columns_to_drop=['approxdate','eventid','related']
df.drop(columns_to_drop,axis=1,inplace=True,)
```

```python
print('Most effect Regions : ',df.region_txt.value_counts().index[0])
print("Most effect Countries form torrorism Attects : ",df.country_txt.value_counts().idxmax())
print("Most effect City form attack : " , df.city.value_counts().index[1])
print('Most terrorist group involved in attacks : ', df.gname.value_counts().index[1])
print("Totel number of killed people : " ,df.nkill.sum())
print("Year with most of attacks : ", df.iyear.value_counts().idxmax())
print("Most of Attacks Types : ",df.attacktype1_txt.value_counts().index[0])
```

```python
plt.subplots(figsize=(20, 10))
sns.countplot(x='iyear', data=df)
plt.title('Attacks by years')
plt.ylabel('Attacks')
plt.xticks(rotation=(90))
plt.show()
plt.xticks(rotation=(45))
```

```python
plt.subplots(figsize=(20, 10))
sns.countplot(x='imonth', data=df)
plt.title('Attacks by months')
plt.ylabel('Attacks')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='region_txt',data=df)
plt.xticks(rotation=(20))
plt.show()
```
