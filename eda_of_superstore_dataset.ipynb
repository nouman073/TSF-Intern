```python
# Importing the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import the dataset
df=pd.read_csv("SampleSuperstore.csv")
```

```python
sns.pairplot(df,hue='Category')
```

```python
sns.scatterplot(df, x='Sales',y='Profit',hue='Category')
```

```python
sns.countplot(x='Category', data=df)
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x='Sub-Category', data=df)
```

```python
plt.figure(figsize=(20,8))
plt.bar('Sub-Category', 'Category', data=df)
```

```python
category_sales_profit = df.groupby(['Category', 'Sub-Category']).agg({'Sales': 'sum', 'Profit': 'sum'})
category_sales_profit = category_sales_profit.sort_values(by='Profit', ascending=False)
print(category_sales_profit.head(20))
```
